```
from urllib.request import urlretrieve
import matplotlib.pyplot as plt
import numpy as np
import cv2
import requests
```

```
url = 'https://www.scienceforums.net/uploads/monthly_2017_09/image.png.4fc6e4dcb770c111e6a6a03f7e3f9d0d.png'
urlretrieve(url, 'physicists.png')

img = cv2.imread('physicists.png', 1)

plt.figure(figsize=(10,6))
plt.imshow(img)
plt.show()
```

```
"""Default"""

url = 'https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml'
urlretrieve(url, 'haarcascade_frontalface_default.xml')

hc  = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
ret = hc.detectMultiScale(img, minNeighbors=1, minSize=(1,1))

img_bbox = img.copy()

for x, y, w, h in ret:
	cv2.rectangle(img_bbox, (x,y), (x+w,y+h), (0,0,255), 2)

plt.figure(figsize=(10,6))
plt.imshow(img_bbox)
plt.show()
```

```
"""Alt tree"""

url = 'https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_alt_tree.xml'
urlretrieve(url, 'haarcascade_frontalface_alt_tree.xml')

hc  = cv2.CascadeClassifier('haarcascade_frontalface_alt_tree.xml')
ret = hc.detectMultiScale(img, minNeighbors=1, minSize=(1,1))

img_bbox = img.copy()

for x, y, w, h in ret:
	cv2.rectangle(img_bbox, (x,y), (x+w,y+h), (0,0,255), 2)

plt.figure(figsize=(10,6))
plt.imshow(img_bbox)
plt.show()
```

```
"""Alt2"""

url = 'https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_alt2.xml'
urlretrieve(url, 'haarcascade_frontalface_alt2.xml')

hc  = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
ret = hc.detectMultiScale(img, minNeighbors=1, minSize=(1,1))

img_bbox = img.copy()

for x, y, w, h in ret:
	cv2.rectangle(img_bbox, (x,y), (x+w,y+h), (0,0,255), 2)

plt.figure(figsize=(10,6))
plt.imshow(img_bbox)
plt.show()
```

```
"""Single face"""

url = 'https://cdn.cnn.com/cnnnext/dam/assets/201025174835-a-chess-prodigy-story-in-the-queens-gambit-00012909-exlarge-169.jpg'
urlretrieve(url, 'Harmon.jpg')

img = cv2.imread('Harmon.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,6))
plt.imshow(img)
plt.show()
```

```
hc   = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
bbox = hc.detectMultiScale(img, minNeighbors=1, minSize=(100,100))
img_bbox = img.copy()

for x, y, w, h in bbox:
	cv2.rectangle(img_bbox, (x,y), (x+w,y+h), (0,0,255), 2)

# img_bbox = cv2.cvtColor(img_bbox, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,6))
plt.imshow(img_bbox)
plt.show()
```

```
"""Crop"""

# select the largest bbox as candidate
area = 0
for x, y, w, h in bbox:
  if w*h > area:
    area = w*h
    (x_, y_, w_, h_) = (x, y, w, h)

    face_img = img[y_:y_+h_, x_:x_+w_]

plt.imshow(face_img)
plt.show()
```
