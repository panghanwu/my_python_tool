```
import numpy as np

size = 500
x = np.random.normal(0, .25, size=size)
y = np.random.normal(0, .25, size=size)
rgb = (np.random.randint(0,1), np.random.randint(0,1), np.random.randint(0,1))

plt.figure(figsize=(10,10))
for i in range(size):
    rgb = (np.random.rand(), np.random.rand(), np.random.rand())
    plt.arrow(.0, .0, x[i], y[i], width=.015, color=rgb)

plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.axis(False)
plt.grid()

plt.show()
```
