```
from urllib.request import urlretrieve
import matplotlib.pyplot as plt
import numpy as np
import cv2

url = 'https://static.wikia.nocookie.net/netflix/images/c/cf/The_Queen%27s_Gambit_S1_Poster_01.jpg'
urlretrieve(url, 'The Queen’s Gambit.jpg')

img = cv2.imread('The Queen’s Gambit.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

```
# prepare square padding
(y, x, rgb) = img.shape

x_pad = 0
y_pad = 0

if x > y:
    y_pad = int((x-y)/2)
else:
    x_pad = int((y-x)/2)

print((x_pad,y_pad))
```

```
"""Constant"""

result = np.pad(img, [(0,0),(x_pad,x_pad),(y_pad,y_pad)], 'constant')

plt.figure(figsize=(6,6))
plt.imshow(result)
plt.show()
```

```
"""Edge"""

result = np.pad(img, [(0,0),(x_pad,x_pad),(y_pad,y_pad)], 'edge')

plt.figure(figsize=(6,6))
plt.imshow(result)
plt.show()
```
